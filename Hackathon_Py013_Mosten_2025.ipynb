```python
import pandas as pd
import plotly.express as px
```

```python
frota_veiculos = "Frota_veiculos_2024.csv"
numero_acidentes_moto = "numero-de-acidentes-envolvendo-moto.csv"

df = pd.read_csv(frota_veiculos, sep=";", encoding="latin1")
df.columns = [col.encode("latin1").decode("utf-8") if isinstance(col, str) else col for col in df.columns]
df['Indicador'] = df['Indicador'].str.encode("latin1").str.decode("utf-8")

df
```

```python
df = df[df['Nível'].str.match(r'1\.1\.\d+')]

colunas_anos = ['2018','2019','2020','2021','2022','2023','2024']
df = df[['Indicador'] + colunas_anos]
df
```

```python
df_acidentes = pd.read_csv(numero_acidentes_moto)
```

```python
# df_acidentes = df_acidentes.drop(columns="codigo", inplace=True)
df_acidentes.rename(columns={
    'data':'ano',
}, inplace=True)
df_acidentes['valor'] = df_acidentes['valor'].astype(int)
df_acidentes['ano'] = df_acidentes['ano'].astype(str).str[:4]
df_acidentes['ano'] = df_acidentes['ano'].astype(int)
df_ordenado_acidentes = df_acidentes.sort_values('ano')

df_ordenado_acidentes = df_ordenado_acidentes[df_ordenado_acidentes['ano'] >= 2018]

df_ordenado_acidentes
```

```python
fig_1 = px.line(df_ordenado_acidentes,
                x='ano',
                y='valor',
                markers=True,
                title='Quantidade de Acidentes de Moto por Ano')
fig_1.update_layout(xaxis_title='Ano', yaxis_title='Quantidade de Acidentes')
fig_1.show()
```

```python
motocicletas = df.iloc[9]
motonetas = df.iloc[10]

motos = pd.DataFrame()

motos['Ano'] = colunas_anos
motos['Quantidade'] = motocicletas[colunas_anos].values.astype(int) + motonetas[colunas_anos].values.astype(int)
motos['Indicador'] = 'Motos'
motos
```

```python
df_acidentes_plot = df_ordenado_acidentes.copy()
df_acidentes_plot.rename(columns={'valor': 'Quantidade'}, inplace=True)
df_acidentes_plot['Indicador'] = 'Acidentes'

df_motos_acidentes = pd.concat([df_acidentes_plot[['ano', 'Quantidade', 'Indicador']], motos[['Ano', 'Quantidade', 'Indicador']].rename(columns={'Ano': 'ano'})])

fig_compare = px.line(df_motos_acidentes,
                      x='ano',
                      y='Quantidade',
                      color='Indicador',
                      markers=True,
                      title='Quantidade de Motos e Acidentes por Ano')
fig_compare.update_layout(xaxis_title='Ano', yaxis_title='Quantidade')
fig_compare.show()
```

```python
df_frota = df.melt(id_vars='Indicador', var_name='Ano', value_name='Quantidade')
df_frota['Ano'] = df_frota['Ano'].astype(int)
df_frota['Quantidade'] = df_frota['Quantidade'].fillna(0).astype(int)
df_frota
```

```python
fig_2 = px.line(df_frota,
                x='Ano',
                y='Quantidade',
                color='Indicador',
                markers=True,
                title='Quantidade de Veículos por Ano')
fig_2.update_layout(xaxis_title='Ano', yaxis_title='Quantidade de Veículos')
fig_2.show()
```

```python
df_pivot = df_frota.pivot(index='Indicador', columns='Ano', values='Quantidade').reset_index()

df_pivot['Crescimento_18_19'] = ((df_pivot[2019] - df_pivot[2018]) / df_pivot[2018]) * 100
df_pivot['Crescimento_19_20'] = ((df_pivot[2020] - df_pivot[2019]) / df_pivot[2019]) * 100
df_pivot['Crescimento_20_21'] = ((df_pivot[2021] - df_pivot[2020]) / df_pivot[2020]) * 100
df_pivot['Crescimento_21_22'] = ((df_pivot[2022] - df_pivot[2021]) / df_pivot[2021]) * 100
df_pivot['Crescimento_22_23'] = ((df_pivot[2023] - df_pivot[2022]) / df_pivot[2022]) * 100
df_pivot['Crescimento_23_24'] = ((df_pivot[2024] - df_pivot[2023]) / df_pivot[2023]) * 100

df_growth = df_pivot.melt(id_vars='Indicador',
                          value_vars=[
                              'Crescimento_18_19',
                              'Crescimento_19_20',
                              'Crescimento_20_21',
                              'Crescimento_21_22',
                              'Crescimento_22_23',
                              'Crescimento_23_24'],
                          var_name='Período', value_name='Crescimento (%)')

df_growth['Período'] = df_growth['Período'].map({
    'Crescimento_18_19': '2018 → 2019',
    'Crescimento_19_20': '2019 → 2020',
    'Crescimento_20_21': '2020 → 2021',
    'Crescimento_21_22': '2021 → 2022',
    'Crescimento_22_23': '2022 → 2023',
    'Crescimento_23_24': '2023 → 2024',
})
```

```python
fig_3 = px.bar(df_growth,
              x="Indicador", y="Crescimento (%)", color="Período", barmode="group",
              title="Crescimento Percentual da Frota por Tipo de Veículo")
fig_3.update_layout(xaxis_title="Tipo de Veículo", yaxis_title="Crescimento (%)")
fig_3.show()
```
